```python
import pandas as pd
from ast import literal_eval
%matplotlib inline
```

```python
df=pd.read_csv('high_volume.csv')
df['issn']=df.issn.apply(literal_eval)
df.shape
```

```python
doaj=pd.read_csv('doaj_20170722_1530_utf8.csv')
doaj=doaj[doaj['APC amount']>0]
doaj_issnp=doaj['Journal ISSN (print version)'].values
doaj_issne=doaj['Journal EISSN (online version)'].values
doaj_issns=[x for x in doaj_issnp if str(x) != 'nan']+[x for x in doaj_issne if str(x) != 'nan']
doaj.shape
```

```python
doaj_rem=pd.read_csv('DOAJ- journals added and removed - Removed.csv')
doaj_rem=doaj_rem[~doaj_rem.r.str.strip().isin(['Journal is no longer Open Access',
                                                'Journal is not Open Access',
                                               'journal not open access'
])]
doaj_rem_issns=doaj_rem.i.values
doaj_rem.shape
```
