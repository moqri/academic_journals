```python
import pandas as pd
```

```python
df=pd.read_csv('high_volum.csv')
```

```python
with open('pred.txt') as f:
    lines = f.read().splitlines()
preds=[line.split('(')[0].strip() for line in lines]
```

```python
with open('doaj.txt') as f:
    lines = f.read().splitlines()
drem=[line.strip() for line in lines]
```

```python
df['In Beall List']= df['5'].apply(lambda x:x in preds)
df['Removed from DOAJ']= df['5'].apply(lambda x:x in drem)
df['Listed']=df['In Beall List']|df['Removed from DOAJ']
```

```python
df['impact']=df['cited']/df['count']
```

```python
df[df['Listed']].sort('impact')
```

```python
hvli=df[df['Listed']][df.impact<.1].sort('impact')[['5','count','cited','impact','In Beall List','Removed from DOAJ']]
hvli.columns=['Journal Name','Count of Articles (2014 to 2016)','Cited Articles','Ratio of Cited Articles','In Beall List','Removed from DOAJ']
hvli
```

```python
hvli.to_csv('high_volume_and_low_impact.csv',index=False)
```
